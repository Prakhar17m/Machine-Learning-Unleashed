```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
```

```python
X , y = make_blobs(n_samples=200 , centers=3 , cluster_std=2 , random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
single_model = AgglomerativeClustering(n_clusters=3 , linkage='single')
```

```python
single_labels = single_model.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=single_labels,cmap='viridis')
```

```python
complete_model = AgglomerativeClustering(n_clusters=3 , linkage='complete')

complete_labels = complete_model.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=complete_labels,cmap='viridis')
```

```python
average_model = AgglomerativeClustering(n_clusters=3 , linkage='average')

average_labels = average_model.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=average_labels,cmap='viridis')
```

```python
ward_model = AgglomerativeClustering(n_clusters=3 , linkage='ward')

ward_labels = ward_model.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=ward_labels,cmap='viridis')
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage
```

```python
ward_linkage = linkage(X , method='ward')
```

```python
dendrogram(ward_linkage)
```
