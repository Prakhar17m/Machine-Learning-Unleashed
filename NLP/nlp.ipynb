```python
import nltk
from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
words = [
    "running", "runs", "ran", "studies", "studying",
    "better", "best", "cars", "children", "mice",
    "flying", "flies", "went", "playing", "played",
    "happier", "happiness"
]
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
for word in words:
  stem = stemmer.stem(word)
  lemma = lemmatizer.lemmatize(word)
  print(f"{word} | {stem} | {lemma}")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
sentences = [
    "love movie",
    "movie great",
    "hate movie",
    "movie terrible",
    "love great acting",
    "hate terrible acting"
]
```

```python
vocab = sorted(set(" ".join(sentences).split()))
```

```python
vocab
```

```python
#this will convert a word to vector and this multiple words vector will create a matrix

import numpy as np
onehot = OneHotEncoder()
onehot.fit(np.array(vocab).reshape(-1,1))

```

```python
for s in sentences:
  words = s.split()
  sentence_matrix = onehot.transform(np.array(words).reshape(-1,1))
  print(sentence_matrix)
```

```python
#Sentences --> vectors
#Bag of words
bow = CountVectorizer()
bow_matrix = bow.fit_transform(sentences)
```

```python
print(bow.get_feature_names_out())
```

```python
print(bow_matrix.toarray())
```

```python
tfidf = TfidfVectorizer(vocabulary=vocab)
tfidf_matrix = tfidf.fit_transform(sentences)
```

```python
print(tfidf.get_feature_names_out())
```

```python
print(tfidf_matrix.toarray())
```
